```julia
using Revise
```

```julia
using ITensors
```

```julia
i, j, k, l = Index(2, "i"), Index(2, "j"), Index(2, "k"), Index(2, "l")
A = random_itensor(i, j, k, l)
Linds = (i, k)
Rinds = (j, l)
D, U = eigen(A, Linds, Rinds)
;
```

```julia
D
U
;
```

```julia
uniqueind(D, U)
```

```julia
dl, dr = uniqueind(D, U), commonind(D, U)
@show dl
@show dr
```

```julia
dl, dr = uniqueind(D, U), commonind(D, U)
Ul = replaceinds(U, (Rinds..., dr) => (Linds..., dl))
```

```julia
Ul = replaceinds(U, (Rinds..., dr) => (Linds..., dl))
```

```julia
A * U ≈ Ul * D # true
```

```julia
@show U * Ul
```
